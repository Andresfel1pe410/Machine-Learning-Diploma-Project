Agregamos la columna CIE_10 GRUPO con las siguientes condiciones.

La clasificación de causas se realizó mediante una función que agrupa los códigos CIE-10 del capítulo I00–I99 en tres categorías epidemiológicas relevantes para el análisis de mortalidad por eventos neurocardiovasculares:

- Isquémicas (I20–I25)

- Hipertensivas (I10–I15)

- Cerebrovasculares (I60–I69)

```python
pipeline = [
    {
        "$addFields": {
            "CIE10_GRUPO": {
                "$switch": {
                    "branches": [
                        {
                            # ISQUÉMICAS I20-I25
                            "case": {
                                "$and": [
                                    {"$gte": [{"$toInt": {"$substr": ["$CIE10_BASICA", 1, 2]}}, 20]},
                                    {"$lte": [{"$toInt": {"$substr": ["$CIE10_BASICA", 1, 2]}}, 25]},
                                    {"$eq": [{"$substr": ["$CIE10_BASICA", 0, 1]}, "I"]}
                                ]
                            },
                            "then": "ISQUEMICAS"
                        },
                        {
                            # CEREBROVASCULARES I60-I69
                            "case": {
                                "$and": [
                                    {"$gte": [{"$toInt": {"$substr": ["$CIE10_BASICA", 1, 2]}}, 60]},
                                    {"$lte": [{"$toInt": {"$substr": ["$CIE10_BASICA", 1, 2]}}, 69]},
                                    {"$eq": [{"$substr": ["$CIE10_BASICA", 0, 1]}, "I"]}
                                ]
                            },
                            "then": "CEREBROVASCULARES"
                        },
                        {
                            # HIPERTENSIVAS I10–I15
                            "case": {
                                "$and": [
                                    {"$gte": [{"$toInt": {"$substr": ["$CIE10_BASICA", 1, 2]}}, 10]},
                                    {"$lte": [{"$toInt": {"$substr": ["$CIE10_BASICA", 1, 2]}}, 15]},
                                    {"$eq": [{"$substr": ["$CIE10_BASICA", 0, 1]}, "I"]}
                                ]
                            },
                            "then": "HIPERTENSIVAS"
                        }
                    ],
                    "default": "OTRAS"
                }
            }
        }
    }
]

update_pipeline = pipeline + [
    {"$merge": {
        "into": "cronicas_ecv",
        "on": "_id",
        "whenMatched": "merge",
        "whenNotMatched": "insert"
    }}
]

collection.aggregate(update_pipeline)
```

Esto se hace con aggregations y análisis cruzado usando pandas.

```python
import matplotlib.pyplot as plt
import pandas as pd

def Graficar(res, nameVariables, variables,title):
    df = pd.DataFrame([
        {
            f"{nameVariables[0]}": item["_id"][f"{variables[0]}"],
            f"{nameVariables[1]}": item["_id"][f"{variables[1]}"],
            f"{nameVariables[2]}": item[f"{variables[2]}"]
        }
        for item in res
    ])

    tabla = df.pivot_table(
        index=nameVariables[0],
        columns=nameVariables[1],
        values=nameVariables[2],
        fill_value=0
    )
    tabla


    plt.figure(figsize=(14, 7))
    tabla.plot(kind='bar', figsize=(15,7))

    plt.title(title)
    plt.xlabel(nameVariables[0])
    plt.ylabel("Número de muertes")
    plt.xticks(rotation=45, ha="right")

    plt.legend(title="CIE-10 GRUPO")
    plt.tight_layout()
    plt.show()
```

```python
pipeline = [
    {"$group": {
        "_id": {"loc": "$LOCALIDAD", "grupo": "$CIE10_GRUPO"},
        "total": {"$sum": 1}
    }},
    {"$sort": {"_id.loc": 1, "total": -1}}
]

res = list(collection.aggregate(pipeline))
res

Graficar(res,
         ["LOCALIDAD","CIE10_GRUPO", "total"],
         ["loc","grupo","total"],
         "Muertes por grupo CIE-10 en cada localidad")
```

```python
pipeline = [
    {"$group": {
        "_id": {"edad": "$EDAD_QUINQUENAL", "grupo": "$CIE10_GRUPO"},
        "total": {"$sum": 1}
    }},
    {"$sort": {"_id.edad": 1}}
]

res = list(collection.aggregate(pipeline))
res

Graficar(res,
         ["EDAD_QUINQUENAL","CIE10_GRUPO", "total"],
         ["edad","grupo","total"],
         "Muertes por grupo CIE-10 en edad quinquenal")
```
