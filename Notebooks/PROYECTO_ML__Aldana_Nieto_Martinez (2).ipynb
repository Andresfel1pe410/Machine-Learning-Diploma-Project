```python
from scipy.stats import chi2_contingency

def cramers_v(confusion_matrix):
    """Calcula el estadístico de Cramér’s V"""
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1)*(r - 1)) / (n - 1))
    rcorr = r - ((r - 1)**2) / (n - 1)
    kcorr = k - ((k - 1)**2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

def chi2_cramersv_all(df, target):
    """Aplica Chi-cuadrado y Cramér’s V a todas las variables categóricas"""
    resultados = []
    cat_cols = df.select_dtypes(include=['object', 'category']).columns

    for col in cat_cols:
        if col == target:
            continue  # No comparar la variable objetivo consigo misma
        tabla = pd.crosstab(df[col], df[target])
        chi2, p, dof, expected = chi2_contingency(tabla)
        v = cramers_v(tabla)
        resultados.append({'Variable': col, 'Chi2': chi2, 'p-value': p, "Cramér's V": v})

    resultados_df = pd.DataFrame(resultados)
    resultados_df = resultados_df.sort_values(by="Cramér's V", ascending=False).reset_index(drop=True)
    return resultados_df

resultados_categoricas = chi2_cramersv_all(df, 'NObeyesdad')
resultados_categoricas

```

El análisis de la asociación entre las variables categóricas y la variable objetivo, revela que todas las variables analizadas son estadísticamente significativas ($p\text{-value} \ll 0.05$). Sin embargo, la fuerza de la asociación varía notablemente, siendo el historial familiar de sobrepeso (family\_history\_with\_overweight) y el género (Gender) los predictores más influyentes, ambos con un coeficiente V de Cramér superior a 0.54, indicando una relación fuerte o muy notable. Por el contrario, factores como el medio de transporte (MTRANS) y el consumo de tabaco (SMOKE) tienen la asociación más débil (V de Cramér $\approx 0.11-0.17$), sugiriendo que, si bien están relacionados, su impacto es menor. En resumen, el modelo debería priorizar la información sobre el historial familiar y el género, ya que demuestran el mayor poder explicativo sobre la variable objetivo.

Método estadístico: el cálculo del IQR permitió detectar de forma numérica los casos extremos por variable.

```python

#variables numéricas ---
num_cols = df.select_dtypes(include=[np.number]).columns
df_num = df[num_cols]

print("\nVARIABLES NUMÉRICAS:")
print(list(num_cols))

#estadistica general ---
print("\nRESUMEN ESTADÍSTICO:")
print(df_num.describe())

# valores atípicos (IQR) ---
outliers_info = {}

for col in num_cols:
    q1 = df_num[col].quantile(0.25)
    q3 = df_num[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = df_num[(df_num[col] < lower) | (df_num[col] > upper)][col]
    outliers_info[col] = {
        "IQR": round(iqr, 3),
        "Límite inferior": round(lower, 3),
        "Límite superior": round(upper, 3),
        "Cantidad atípicos": len(outliers)
    }

#resumen de outliers
outliers_df = pd.DataFrame(outliers_info).T
print("\nVALORES ATÍPICOS DETECTADOS (Método IQR):")
print(outliers_df)

#Boxplots
import matplotlib.pyplot as plt

df_num.boxplot(figsize=(10,6))
plt.title("Detección visual de valores atípicos (Boxplot)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **6. Entrenamiento del Modelo y Selección de Hiperparámetros**
---

Entrene el modelo y consiga la combinación eficiente de hiperparámetros para obtener un rendimiento óptimo.

Es importante visualizar los resultados de manera clara y comprensible. Para esto, **se deben utilizar tablas y gráficos** que muestren cómo cambian las métricas de desempeño del proceso de entrenamiento en función de los hiperparámetros probados. Algunas herramientas visuales útiles incluyen las *curvas de aprendizaje* y las *curvas de validación,* que nos permiten comparar el rendimiento del modelo con diferentes ajustes de hiperparámetros.

```python
# ==========================================
# PUNTO 6 – Entrenamiento del modelo y selección de hiperparámetros
# Comparación de 7 clasificadores con GridSearchCV + Curvas de validación
# ==========================================
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt
from IPython.display import display

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline

# Modelos
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier



TARGET = "NObeyesdad"
X = df.drop(columns=[TARGET])
y = df[TARGET]

cat_cols = X.select_dtypes(include="object").columns.tolist()
num_cols = X.select_dtypes(exclude="object").columns.tolist()

pre = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), cat_cols),
        ("num", StandardScaler(), num_cols),
    ],
    remainder="drop"
)

# Split base (la métrica que miramos es de CV; este split se usa para refit)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, stratify=y, random_state=42
)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = "f1_macro"

# 2) Definir modelos y grids de hiperparámetros
models_and_grids = {
    "Logistic Regression": (
        LogisticRegression(max_iter=1000, multi_class="multinomial"),
        {
            "model__C": [0.1, 1, 10],
            "model__penalty": ["l2"],
            "model__solver": ["lbfgs"],
        }
    ),
    "Gaussian Naive Bayes": (
        GaussianNB(),
        {
            "model__var_smoothing": [1e-9, 1e-8, 1e-7]
        }
    ),
    "KNN": (
        KNeighborsClassifier(),
        {
            "model__n_neighbors": [3, 5, 7, 9, 11],
            "model__weights": ["uniform", "distance"],
            "model__p": [1, 2],  # 1=Manhattan, 2=Euclidiana
        }
    ),
    "Decision Tree": (
        DecisionTreeClassifier(random_state=42),
        {
            "model__max_depth": [None, 5, 10, 20],
            "model__min_samples_leaf": [1, 2, 5],
            "model__criterion": ["gini", "entropy"],
        }
    ),
    "Random Forest": (
        RandomForestClassifier(random_state=42, n_jobs=-1),
        {
            "model__n_estimators": [100, 300, 500],
            "model__max_depth": [None, 10, 20],
            "model__min_samples_leaf": [1, 2],
        }
    ),
    "SVM": (
        SVC(probability=True, random_state=42),
        {
            "model__C": [0.1, 1, 10],
            "model__kernel": ["rbf"],
            "model__gamma": ["scale", "auto"],
        }
    ),
    "RedNeuronal": (
        MLPClassifier(max_iter=500, random_state=42),
        {
            "model__hidden_layer_sizes": [(64,), (128,), (64, 32)],
            "model__alpha": [1e-4, 1e-3],
            "model__learning_rate_init": [1e-3, 5e-3],
        }
    ),
}

# 3) Ejecutar GridSearchCV para cada modelo

all_results = []          # resumen comparativo
per_model_cv = {}         # cv_results_ por modelo para curvas
best_pipes = {}           # mejor pipeline por modelo

for name, (estimator, grid) in models_and_grids.items():
    pipe = Pipeline(steps=[("pre", pre), ("model", estimator)])
    gs = GridSearchCV(
        estimator=pipe,
        param_grid=grid,
        scoring=scoring,
        cv=cv,
        n_jobs=-1,
        refit=True,
        return_train_score=True,
        verbose=0
    )
    gs.fit(X_train, y_train)

    per_model_cv[name] = pd.DataFrame(gs.cv_results_)
    best_pipes[name] = gs.best_estimator_

    all_results.append({
        "Modelo": name,
        "Mejor F1-macro (CV)": gs.best_score_,
        "Mejores parámetros": gs.best_params_
    })

results_df = pd.DataFrame(all_results).sort_values("Mejor F1-macro (CV)", ascending=False)
print("RESULTADOS COMPARATIVOS (Validación Cruzada, métrica: F1-macro):")
display(results_df)

# 4) Curvas de validación por hiperparámetro clave
def plot_validation_curve(cv_df, param_name, title, x_as_str=False):
    """Grafica mean_test_score vs un hiperparámetro.
       Si x_as_str=True, convierte los valores de ese hiperparámetro a str (para tuplas como hidden_layer_sizes)."""
    if param_name not in cv_df.columns:
        print(f"[Aviso] {param_name} no está en este grid.")
        return
    tmp = cv_df.copy()
    if x_as_str:
        tmp[param_name] = tmp[param_name].astype(str)
    grouped = tmp.groupby(param_name, as_index=False)["mean_test_score"].mean()
    # orden razonable del eje X
    try:
        x_vals = grouped[param_name].astype(float)
        order_idx = np.argsort(x_vals.values)
    except Exception:
        order_idx = np.arange(len(grouped))
    plt.figure(figsize=(7,4))
    plt.plot(grouped.iloc[order_idx, 0], grouped.iloc[order_idx]["mean_test_score"], marker="o")
    plt.xlabel(param_name.replace("param_model__", ""))
    plt.ylabel("F1-macro (CV)")
    plt.title(title)
    plt.grid(True, alpha=.3)
    plt.tight_layout()
    plt.show()

# Logistic Regression
plot_validation_curve(per_model_cv["Logistic Regression"], "param_model__C",
                      "Logistic Regression: C vs F1-macro (CV)")

# GaussianNB
plot_validation_curve(per_model_cv["Gaussian Naive Bayes"], "param_model__var_smoothing",
                      "GaussianNB: var_smoothing vs F1-macro (CV)")

# KNN
plot_validation_curve(per_model_cv["KNN"], "param_model__n_neighbors",
                      "KNN: n_neighbors vs F1-macro (CV)")

# Decision Tree
plot_validation_curve(per_model_cv["Decision Tree"], "param_model__max_depth",
                      "Decision Tree: max_depth vs F1-macro (CV)")

# Random Forest
plot_validation_curve(per_model_cv["Random Forest"], "param_model__n_estimators",
                      "Random Forest: n_estimators vs F1-macro (CV)")
plot_validation_curve(per_model_cv["Random Forest"], "param_model__max_depth",
                      "Random Forest: max_depth vs F1-macro (CV)")

# SVM
plot_validation_curve(per_model_cv["SVM"], "param_model__C",
                      "SVM (RBF): C vs F1-macro (CV)")

# RE Neural
plot_validation_curve(per_model_cv["RedNeuronal"], "param_model__hidden_layer_sizes",
                      "MLP: hidden_layer_sizes vs F1-macro (CV)", x_as_str=True)

```

- **Regresión**: $r^2$, error cuadrático medio, error absoluto medio, etc.
- **Agrupamiento**: coeficiente de silueta, índice de Davies-Bouldin, etc.

Los resultados se deben presentar usando tablas y figuras además del análisis detallado respectivo.

Los resultados obtenidos en la fase de evaluación permiten comparar el rendimiento de los siete clasificadores implementados en términos de Accuracy y F1-macro sobre el conjunto de prueba.

Como se observa, los modelos con mejor desempeño general fueron el SVM con kernel RBF (Accuracy = 0.9432, F1-macro = 0.9419) y la Red Neuronal (MLPClassifier) (Accuracy = 0.9401, F1-macro = 0.9382). Ambos mostraron una excelente capacidad para generalizar y distinguir entre las diferentes categorías de obesidad.

El Árbol de Decisión también alcanzó un rendimiento notable (F1-macro = 0.9331), lo que sugiere que la estructura jerárquica del modelo logra capturar patrones interpretables en los datos.
La Regresión Logística y el Random Forest mantuvieron resultados sólidos (con F1-macro cercanos a 0.92), confirmando su estabilidad y buen ajuste bajo un esquema multiclase.

El Naive Bayes Gaussiano obtuvo un desempeño significativamente menor (F1-macro = 0.47), lo que indica que las fuertes correlaciones entre las variables del dataset violan su supuesto de independencia.

En conjunto, estos resultados evidencian que los métodos basados en aprendizaje no lineal (SVM y MLP) logran una mejor representación de la compleja relación entre hábitos alimenticios, factores físicos y niveles de obesidad.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay

# Diccionario para guardar métricas
metrics = {}

# Evaluar cada modelo
for name, pipe in best_pipes.items():
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='macro')  # para multiclase
    metrics[name] = {'accuracy': acc, 'f1_macro': f1}

    print(f"\n=== {name} ===")
    print(f"Accuracy: {acc:.4f}")
    print(f"F1-macro: {f1:.4f}")

    # Mostrar matriz de confusión
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues')
    plt.title(f"Matriz de confusión - {name}")
    plt.show()

# Graficar comparación de métricas
names = list(metrics.keys())
accs = [metrics[n]['accuracy'] for n in names]
f1s = [metrics[n]['f1_macro'] for n in names]

x = range(len(names))
plt.figure(figsize=(8,5))
plt.bar(x, accs, width=0.4, label='Accuracy')
plt.bar([i + 0.4 for i in x], f1s, width=0.4, label='F1-macro')
plt.xticks([i + 0.2 for i in x], names, rotation=45)
plt.ylabel("Score")
plt.title("Comparación de desempeño de modelos multiclase")
plt.legend()
plt.tight_layout()
plt.show()
```
